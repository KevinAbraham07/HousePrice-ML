```python
import pandas as pd
```

```python
df=pd.read_csv('Housing.csv')
df.head()
```

```python
df.describe()
```

```python
df.head()
```

```python
list1=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
df[list1]=df[list1].replace({'yes':1,'no':0})
```

```python
df.head()
```

```python
df['furnishingstatus']=df['furnishingstatus'].replace({'furnished':0,'semi-furnished':1,'unfurnished':2})
df.tail()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
list2=['price','area']
df[list2]=scaler.fit_transform(df[list2])
df.head()
```

```python
core=df.corr()
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,3))
sns.heatmap(core,annot=True,cmap='coolwarm')
plt.show()
```

```python
df.hist(figsize=(10,10),bins=10)
plt.show()
```

```python
X=df.drop('price',axis=1)
X
```

```python
y=df['price']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=44)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_predict=lr.predict(X_test)
from sklearn.metrics import r2_score
lr_accuracy=r2_score(y_test,y_predict)
lr_accuracy
```
